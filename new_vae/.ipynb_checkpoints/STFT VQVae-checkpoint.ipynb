```python
train_dataset, test_dataset = get_training_set()
```

```python
vqvae_trainer = VQVAETrainer(latent_dim=256, num_embeddings=128)
vqvae_trainer.load_weights("/home/jerms/daniil/new_vae/saved_models/vqvae_run_stft_5")
vqvae_trainer.compile(optimizer=keras.optimizers.Adam())
```

```python
def get_predictions(model, test_samples):
    predictions = model.vqvae(test_samples)
    return predictions
```

```python
for test_batch in test_dataset.take(1):
    test_sample = test_batch[0][0:16, :, :, :]
```

```python
preds = get_predictions(vqvae_trainer, test_sample)
```

```python
db_stft = test_sample[audio_index]
real_comp = db_stft[:, :, 0].numpy() * 120
```
