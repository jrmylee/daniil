```python
codes_dir = "/Users/llewyn/Documents/data/stft/codes"
sample = np.load(os.path.join(codes_dir,"MIDI-Unprocessed_Recital1-3_MID--AUDIO_02_R1_2018_wav--1.wav-18.npy"), allow_pickle=True)
```

```python
def load_codes(index=0):
    files = os.listdir(codes_dir)[:1000]
    top_samples = [np.load(os.path.join(codes_dir, f), allow_pickle=True)[index] for f in files]
    labels = [f[:f.index(".wav")] for f in files]
    return np.array(top_samples), np.array(labels)
```

```python
top_codes, labels = load_codes(0)
bot_codes, labels = load_codes(1)
```

```python
sns.set_style('darkgrid')

mds = MDS(n_components=2, metric=True, random_state=2)
X_mds = mds.fit_transform(top_codes)

sns.scatterplot(X_mds[:,0], X_mds[:,1],hue=labels, palette='Set1', legend = False)
plt.title("MDS of IRIS dataset", fontsize=15, pad=15)
```

```python
sns.set_style('darkgrid')

mds = MDS(n_components=2, metric=True, random_state=2)
X_mds = mds.fit_transform(bot_codes)

sns.scatterplot(X_mds[:,0], X_mds[:,1],hue=labels, palette='Set1', legend = False)
plt.title("MDS of IRIS dataset", fontsize=15, pad=15)
```
